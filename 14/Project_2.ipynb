### Описание проекта: промышленность
Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Мне предстоит построить модель, которая предскажет температуру стали. 
#### Описание этапа обработки 
Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша.

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»). 

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки. 

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»). 

#### Описание данных 
Данные состоят из файлов, полученных из разных источников: 
- data_arc_new.csv — данные об электродах; 
- data_bulk_new.csv — данные о подаче сыпучих материалов (объём); 
- data_bulk_time_new.csv — данные о подаче сыпучих материалов (время); 
- data_gas_new.csv — данные о продувке сплава газом; 
- data_temp_new.csv — результаты измерения температуры; 
- data_wire_new.csv — данные о проволочных материалах (объём); 
- data_wire_time_new.csv — данные о проволочных материалах (время).

Во всех файлах столбец key содержит номер партии. В файлах может быть несколько строк с одинаковым значением key: они соответствуют разным итерациям обработки.

#### 5. Обработка отсутствующих значений и выбросов, если таковые имеются, чтобы гарантировать качество моделирования.

```python
data_temp_new.dropna(subset=['Температура'], inplace=True)
data_bulk_new.fillna(0, inplace=True)
data_bulk_time_new.fillna(0, inplace=True)
data_wire_new.fillna(0, inplace=True)
data_wire_time_new.fillna(0, inplace=True)
```

```python
# Анализ данных
# Пример анализа для файла data_temp_new.csv
temperature = data_temp_new['Температура'] # Замените 'Температура' на соответствующее имя столбца с температурой в вашем файле
mean = temperature.mean()
std = temperature.std()
threshold = mean + 3 * std # Задаем пороговое значение, например, 3 стандартных отклонения

# Визуализация
plt.figure(figsize=(10, 6))
plt.hist(temperature, bins=50)
plt.axvline(mean, color='r', linestyle='--', label='Среднее')
plt.axvline(threshold, color='g', linestyle='--', label='Пороговое значение')
plt.xlabel('Температура')
plt.ylabel('Частота')
plt.title('Распределение температуры')
plt.legend()
plt.show()
```

```python
# Анализ данных для столбца "Активная мощность"
active_power = data_arc_new['Активная мощность'] # Замените 'Активная мощность' на соответствующее имя столбца в вашем файле
mean_active = active_power.mean()
std_active = active_power.std()
threshold_active = mean_active + 3 * std_active

# Визуализация
plt.figure(figsize=(10, 6))
plt.hist(active_power, bins=50)
plt.axvline(mean_active, color='r', linestyle='--', label='Среднее')
plt.axvline(threshold_active, color='g', linestyle='--', label='Пороговое значение')
plt.xlabel('Активная мощность')
plt.ylabel('Частота')
plt.title('Распределение активной мощности')
plt.legend()
plt.show()

# Анализ данных для столбца "Реактивная мощность"
reactive_power = data_arc_new['Реактивная мощность'] # Замените 'Реактивная мощность' на соответствующее имя столбца в вашем файле
mean_reactive = reactive_power.mean()
std_reactive = reactive_power.std()
threshold_reactive = mean_reactive + 3 * std_reactive

# Визуализация
plt.figure(figsize=(10, 6))
plt.hist(reactive_power, bins=50)
plt.axvline(mean_reactive, color='r', linestyle='--', label='Среднее')
plt.axvline(threshold_reactive, color='g', linestyle='--', label='Пороговое значение')
plt.xlabel('Реактивная мощность')
plt.ylabel('Частота')
plt.title('Распределение реактивной мощности')
plt.legend()
plt.show()
```

```python
# Удаляем столбец "key"
data_bulk_new_filtered = data_bulk_new.drop(columns=['key'])

# Анализ данных для всех столбцов, кроме "key", в data_bulk_new_filtered
for column in data_bulk_new_filtered.columns:
    data = data_bulk_new_filtered[column]
    mean = data.mean()
    std = data.std()
    threshold = mean + 3 * std
    
    # Визуализация
    plt.figure(figsize=(10, 6))
    sns.boxplot(data)
    plt.axhline(threshold, color='r', linestyle='--', label='Пороговое значение')
    plt.xlabel(column)
    plt.ylabel('Значение')
    plt.title('Boxplot для столбца {}'.format(column))
    plt.legend()
    plt.show()
```

```python
# Анализ данных для столбца 'Газ 1' в data_gas_new
gas_1 = data_gas_new['Газ 1']
mean = gas_1.mean()
std = gas_1.std()
threshold = mean + 3 * std

# Визуализация
plt.figure(figsize=(10, 6))
plt.hist(gas_1, bins=50)
plt.axvline(mean, color='r', linestyle='--', label='Среднее')
plt.axvline(threshold, color='g', linestyle='--', label='Пороговое значение')
plt.xlabel('Газ 1')
plt.ylabel('Частота')
plt.title('Распределение значения Газ 1')
plt.legend()
plt.show()
```

```python
# Анализ данных для всех столбцов в data_wire_new
for column in data_wire_new.columns:
    if column != 'key':
        values = data_wire_new[column]
        mean = values.mean()
        std = values.std()
        threshold = mean + 3 * std

        # Визуализация
        plt.figure(figsize=(10, 6))
        plt.hist(values, bins=50)
        plt.axvline(mean, color='r', linestyle='--', label='Среднее')
        plt.axvline(threshold, color='g', linestyle='--', label='Пороговое значение')
        plt.xlabel(column)
        plt.ylabel('Частота')
        plt.title(f'Распределение значения {column}')
        plt.legend()
        plt.show()
```

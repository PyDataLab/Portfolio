## Обучение моделей

````python
# Определение признаков и целевого признака
features = data[['VehicleType', 'RegistrationYear', 'Gearbox', 'Power', 'Kilometer', 'FuelType', 'Brand', 'Repaired', 'PostalCode']]
target = data['Price']

# Кодирование категориальных признаков
label_encoder = LabelEncoder()
for column in features.columns:
    if features[column].dtype == 'object':
        features[column] = label_encoder.fit_transform(features[column].astype(str))

# Разделение на обучающую и тестовую выборки
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.2, random_state=42)
````

````python
# Создание и обучение модели линейной регрессии
model = LinearRegression()
start_time = time.time()
model.fit(features_train, target_train)
train_time = time.time() - start_time

# Вычисление RMSE на обучающей выборке с помощью кросс-валидации
scores = -cross_val_score(model, features_train, target_train, cv=5, scoring='neg_mean_squared_error')
rmse = np.sqrt(np.mean(scores))

# Подбор лучших гиперпараметров с помощью GridSearchCV
params = {'fit_intercept': [True, False], 'normalize': [True, False]}
grid = GridSearchCV(model, params, cv=5, scoring='neg_mean_squared_error')
start_time = time.time()
grid.fit(features_train, target_train)
best_params = grid.best_params_
best_model = grid.best_estimator_
best_train_time = time.time() - start_time

# Вычисление времени предсказания
start_time = time.time()
predictions = best_model.predict(features_train)
predict_time = time.time() - start_time

print(f"RMSE на обучающей выборке: {round(rmse, 2)}")
print(f"Лучшие гиперпараметры: {best_params}")
print(f"Время обучения: {round(best_train_time, 2)} сек")
print(f"Время предсказания: {round(predict_time, 2)} сек")
````

````python
# Создание модели решающего дерева регрессии
model = DecisionTreeRegressor()

# Задание возможных значений гиперпараметров для подбора
params = {
    'max_depth': [None, 5, 10, 15], # Максимальная глубина дерева
    'min_samples_split': [2, 5, 10], # Минимальное количество образцов для разделения узла
    'min_samples_leaf': [1, 2, 4] # Минимальное количество образцов в листе
}

# Подбор лучших гиперпараметров с использованием GridSearchCV
grid_search = GridSearchCV(model, params, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(features_train, target_train)

# Лучшие гиперпараметры
best_params = grid_search.best_params_

# Создание модели решающего дерева регрессии с лучшими гиперпараметрами
best_model = DecisionTreeRegressor(max_depth=best_params['max_depth'],
                                   min_samples_split=best_params['min_samples_split'],
                                   min_samples_leaf=best_params['min_samples_leaf'])

# Вычисление RMSE с помощью кросс-валидации на тренировочной выборке
rmse_scores = -cross_val_score(best_model, features_train, target_train, cv=5, scoring='neg_mean_squared_error')
rmse = np.sqrt(np.mean(rmse_scores))

# Вычисление времени обучения
start_time = time.time()
best_model.fit(features_train, target_train)
train_time = time.time() - start_time

# Вычисление времени предсказания
start_time = time.time()
predictions = best_model.predict(features_train)
predict_time = time.time() - start_time

print("Лучшие гиперпараметры: ", best_params)
print("RMSE на тренировочной выборке: ", round(rmse,2))
print("Время обучения: ", round(train_time,2), "сек")
print("Время предсказания: ", round(predict_time,2), "сек")
````

````python
# Определение диапазонов значений гиперпараметров
n_estimators = [100, 300]
max_depth = [10, 30]
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]

# Создание словаря с гиперпараметрами
param_grid = {'n_estimators': n_estimators,
'max_depth': max_depth,
'min_samples_split': min_samples_split,
'min_samples_leaf': min_samples_leaf,
'bootstrap': bootstrap}

# Создание модели случайного леса
rf = RandomForestRegressor()

# Подбор лучших гиперпараметров
rf_random = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, n_iter=1, cv=3)
start_time = time.time()
rf_random.fit(features_train, target_train)
end_time = time.time()

# Вывод лучших гиперпараметров
best_params = rf_random.best_params_
print("Лучшие гиперпараметры:", best_params)

# Обучение модели с лучшими гиперпараметрами на тренировочной выборке
best_rf = RandomForestRegressor(**best_params)
start_train_time = time.time()
best_rf.fit(features_train, target_train)
end_train_time = time.time()

# Предсказание на тренировочной выборке
start_pred_time = time.time()
predictions_train = best_rf.predict(features_train)
end_pred_time = time.time()

# Расчет RMSE на тренировочной выборке
rmse_train = sqrt(mean_squared_error(target_train, predictions_train))
print("RMSE на тренировочной выборке:", round(rmse_train,2))

# Время выполнения обучения модели и предсказания на тренировочной выборке
train_time = end_train_time - start_train_time
pred_time = end_pred_time - start_pred_time
print("Время обучения модели на тренировочной выборке:", round(train_time,2), "секунд")
print("Время предсказания на тренировочной выборке:", round(pred_time,2), "секунд")
````

````python
# Определение модели LightGBM
model = lgb.LGBMRegressor()

# Определение диапазонов значений для гиперпараметров
param_grid = {
    'num_leaves': [31, 63, 127],
    'max_depth': [None, 5, 10],
    'learning_rate': [0.05, 0.1, 0.2]
}

# Подбор гиперпараметров с помощью GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_root_mean_squared_error', cv=5)
start_time = time.time()
grid_search.fit(features_train, target_train)
end_time = time.time()

# Вычисление RMSE
best_model = grid_search.best_estimator_
predictions = best_model.predict(features_train)
rmse = mean_squared_error(target_train, predictions, squared=False)

# Вычисление времени обучения и времени предсказания
fit_time = end_time - start_time
predict_time = time.time() - end_time

# Вывод результатов
print("Лучшие гиперпараметры:", grid_search.best_params_)
print("RMSE:", round(rmse,2))
print("Время обучения (сек):", round(fit_time,2))
print("Время предсказания (сек):", round(predict_time,2))
````

````python
# Определение модели CatBoost
model = CatBoostRegressor()

# Определение диапазона значений гиперпараметров для подбора
param_grid = {
    'iterations': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'depth': [4, 6, 8]
}

# Подбор гиперпараметров с использованием GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_root_mean_squared_error')
start_time = time.time()
grid_search.fit(features_train, target_train)
end_time = time.time()

# Вычисление RMSE 
best_model = grid_search.best_estimator_
predictions = best_model.predict(features_train)
rmse = np.sqrt(mean_squared_error(target_train, predictions))

# Вычисление времени обучения и времени предсказания
training_time = end_time - start_time
prediction_time = time.time() - end_time

# Вывод результатов
print('Лучшие гиперпараметры:', grid_search.best_params_)
print('RMSE на тренировочной выборке:', round(rmse,2))
print('Время обучения:', round(training_time,2))
print('Время предсказания:', round(prediction_time,2))
````

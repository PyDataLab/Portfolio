### Описание проекта: промышленность
Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Мне предстоит построить модель, которая предскажет температуру стали. 
#### Описание этапа обработки 
Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша.

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»). 

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки. 

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»). 

#### Описание данных 
Данные состоят из файлов, полученных из разных источников: 
- data_arc_new.csv — данные об электродах; 
- data_bulk_new.csv — данные о подаче сыпучих материалов (объём); 
- data_bulk_time_new.csv — данные о подаче сыпучих материалов (время); 
- data_gas_new.csv — данные о продувке сплава газом; 
- data_temp_new.csv — результаты измерения температуры; 
- data_wire_new.csv — данные о проволочных материалах (объём); 
- data_wire_time_new.csv — данные о проволочных материалах (время).

Во всех файлах столбец key содержит номер партии. В файлах может быть несколько строк с одинаковым значением key: они соответствуют разным итерациям обработки.

## Часть 2: Предобработка данных, разработка модели

```python
# Преобразование столбцов 'Начало нагрева дугой' и 'Конец нагрева дугой' в формат datetime
data_arc_new['Начало нагрева дугой'] = pd.to_datetime(data_arc_new['Начало нагрева дугой'])
data_arc_new['Конец нагрева дугой'] = pd.to_datetime(data_arc_new['Конец нагрева дугой'])

# Расчет времени нагрева в секундах и добавление нового столбца 'Время нагрева в сек'
data_arc_new['Время нагрева в сек'] = (data_arc_new['Конец нагрева дугой'] - 
                                   data_arc_new['Начало нагрева дугой']).dt.seconds

# Создадим датасет, сгруппируем по key и агрегируем его по суммам переменных
data_arc_new = data_arc_new.groupby('key').agg(
    total_heating_time_seconds=('Время нагрева в сек', 'sum'),
    P=('Активная мощность', 'sum'),
    Q=('Реактивная мощность', 'sum')
)

# Найдем полную мощность
data_arc_new['Полная мощность'] = np.sqrt(data_arc_new['P']**2 + data_arc_new['Q']**2)

# Объединение данных из разных источников по ключу 'key'
data = data_temp_new.merge(right=data_arc_new,
                       how='inner',
                       on='key')\
                .merge(right=data_bulk_new,
                       how='inner',
                       on='key')\
                .merge(right=data_gas_new,
                       how='inner',
                       on='key')\
                .merge(right=data_wire_new,
                       how='inner',
                       on='key')

# Удаление столбцов 'Время замера', 'Начало нагрева дугой' и 'Конец нагрева дугой'
merged_data = data.drop('Время замера', axis=1)

# Удаление строк с отсутствующими значениями температуры
merged_data.dropna(subset=['Температура'], inplace=True)

# Заполнение всех остальных пропущенных значений нулями
merged_data = merged_data.fillna(0)

# Подсчет количества строк для каждого значения ключа
key_counts = merged_data.groupby('key').size().reset_index(name='count')

# Определение ключей, для которых есть более одной строки
keys_to_keep = key_counts[key_counts['count'] > 1]['key']

# Оставляем только строки с ключами, для которых есть более одной строки
merged_data = merged_data[merged_data['key'].isin(keys_to_keep)]

# Создание нового столбца 'first_temperature' со значением первой температуры для каждого ключа
merged_data['first_temperature'] = merged_data.groupby('key')['Температура'].transform('first')

# Удаление всех строк с одинаковыми ключами, кроме последней
merged_data = merged_data.drop_duplicates(subset='key', keep='last')

# Сброс индексов
merged_data = merged_data.reset_index(drop=True)

# Вывод результата
merged_data
```

<div class="alert alert-block alert-info">
    
Сначала мы преобразовали столбцы 'Начало нагрева дугой' и 'Конец нагрева дугой' в формат datetime и рассчитали время нагрева в секундах. Затем мы создали новый набор данных, сгруппировав по `key` и агрегировав по суммам переменных.

Далее мы объединили данные из разных источников по `key` и удалили столбцы, которые были не нужны. Мы также удалили строки с отсутствующими значениями температуры и заполнили все остальные пропущенные значения нулями.

Затем мы подсчитали количество строк для каждого значения `key` и оставили только строки с ключами, которые имели более одной строки. Мы создали новый столбец 'first_temperature' со значением первой температуры для каждого ключа и удалили все строки с дублирующимися ключами, кроме последней. Наконец, мы сбросили индексы.
    
</div>

```python
# Разделение данных на обучающую и тестовую выборки
RANDOM_STATE = 30723
train_data, test_data = train_test_split(merged_data, test_size=0.25, random_state=RANDOM_STATE)

# Проверка размерности данных
print("Размер обучающей выборки:", train_data.shape)
print("Размер тестовой выборки:", test_data.shape)
```

```python
# Определение признаков и целевого признака
features = train_data.drop(['Температура', 'key'], axis=1)
target = train_data['Температура']

# Создание списка моделей
models = [LinearRegression(), RandomForestRegressor(random_state=RANDOM_STATE), CatBoostRegressor(random_state=RANDOM_STATE, verbose=False), LGBMRegressor(random_state=RANDOM_STATE)]

# Определение лучшей модели по метрике на кросс-валидации
best_score = float('inf')
best_model = None

for model in models:
    scores = cross_val_score(model, features, target, cv=5, scoring='neg_mean_absolute_error')
    mean_score = abs(scores.mean())
    if mean_score < best_score:
        best_score = mean_score
        best_model = model

print(f'Лучшая модель: {best_model}')
print(f'Средняя абсолютная ошибка на кросс-валидации: {round(best_score,2)}')
print(f'Параметры лучшей модели: {best_model.get_params()}')
```

<div class="alert alert-block alert-info">
    
Мы определили признаки и целевой признак для обучения моделей. Затем мы создали список моделей, включающий линейную регрессию, случайный лес, CatBoost и LightGBM.

Далее мы использовали кросс-валидацию с 5 блоками для оценки качества каждой модели по метрике средней абсолютной ошибки. Мы выбрали лучшую модель как ту, которая имеет наименьшую среднюю абсолютную ошибку на кросс-валидации.

В результате мы нашли лучшую модель CatBoostRegressor, среднюю абсолютную ошибку на кросс-валидации: 6.26 и параметры: {'loss_function': 'RMSE', 'verbose': False, 'random_state': 30723}.
    
</div>

```python
# Определение признаков и целевого признака
features_train = train_data.drop(['Температура', 'key'], axis=1)
target_train = train_data['Температура']
features_test = test_data.drop(['Температура', 'key'], axis=1)
target_test = test_data['Температура']

def objective(trial):
    # Определение гиперпараметров для оптимизации
    params = {
        'iterations': trial.suggest_int('iterations', 50, 300),
        'depth': trial.suggest_int('depth', 4, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'random_strength': trial.suggest_int('random_strength', 0, 100),
        'bagging_temperature': trial.suggest_float('bagging_temperature', 0, 1),
        'od_type': trial.suggest_categorical('od_type', ['IncToDec', 'Iter']),
        'od_wait': trial.suggest_int('od_wait', 10, 50)
    }
    
    # Создание модели CatBoostRegressor с указанными гиперпараметрами
    model = CatBoostRegressor(**params, random_state=RANDOM_STATE, loss_function='RMSE', verbose=False)
    
    # Оценка качества модели с помощью кросс-валидации
    scores = cross_val_score(model, features_train, target_train, cv=5, scoring='neg_mean_absolute_error')
    mean_score = abs(scores.mean())
    
    return mean_score

# Создание экземпляра класса Optuna study для оптимизации гиперпараметров
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)

# Вывод лучших гиперпараметров
print(f'Лучшие гиперпараметры: {study.best_params}')

# Создание модели CatBoostRegressor с лучшими гиперпараметрами
best_params = study.best_params
best_model = CatBoostRegressor(**best_params, random_state=RANDOM_STATE, loss_function='RMSE', verbose=False)

# Обучение модели на обучающей выборке
best_model.fit(features_train, target_train)

# Предсказание температуры на тестовой выборке
predicted_test = best_model.predict(features_test)

# Расчет средней абсолютной ошибки на тестовой выборке
mae_test = mean_absolute_error(target_test, predicted_test)

print(f'Средняя абсолютная ошибка на тестовой выборке: {round(mae_test,2)}')
```

<div class="alert alert-block alert-info">
    
Мы определили признаки и целевой признак для обучения и тестирования моделей. Затем мы использовали библиотеку Optuna для оптимизации гиперпараметров модели CatBoostRegressor. Мы определили гиперпараметры для оптимизации и создали функцию `objective`, которая принимает объект `trial` и возвращает среднюю абсолютную ошибку на кросс-валидации для модели CatBoostRegressor с гиперпараметрами, определенными в `trial`.

Мы создали экземпляр класса `optuna.study` и использовали его для оптимизации гиперпараметров, вызывая метод `optimize` с функцией `objective` и числом испытаний. В результате мы нашли лучшие гиперпараметры и вывели их.

Затем мы создали модель CatBoostRegressor с лучшими гиперпараметрами и обучили ее на обучающей выборке. Мы использовали обученную модель для предсказания температуры на тестовой выборке и рассчитали среднюю абсолютную ошибку на тестовой выборке.

В результате мы получили модель CatBoostRegressor с оптимизированными гиперпараметрами: {'iterations': 267, 'depth': 4, 'learning_rate': 0.08180416023734272, 'random_strength': 0, 'bagging_temperature': 0.6038000917776002, 'od_type': 'IncToDec', 'od_wait': 10}, которая показывает хорошее качество предсказания температуры стали на тестовой выборке: 6.04.

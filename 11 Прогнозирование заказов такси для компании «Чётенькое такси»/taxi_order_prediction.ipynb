#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Подготовка

```python
# Импортируем библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
from sklearn.metrics import mean_squared_error
```

```python
# Загружаем данные
data = pd.read_csv('datasets/taxi.csv')
```

```python
# преобразование столбца с датой и временем к типу datetime
data['datetime'] = pd.to_datetime(data['datetime'])

# установка индекса DataFrame на столбец datetime
data.set_index('datetime', inplace=True) 

# ресемплирование по одному часу и вычисление суммы значений
data = data.resample('1H').sum()
```

## Обучение

```python
def make_features(data, max_lag, rolling_mean_size):
    data['year'] = data.index.year # Добавляем столбец с годом из индекса времени
    data['month'] = data.index.month # Добавляем столбец с месяцем из индекса времени
    data['day'] = data.index.day # Добавляем столбец с днем месяца из индекса времени
    data['dayofweek'] = data.index.dayofweek # Добавляем столбец с номером дня недели (0-6, пн-вс)

    # Добавляем столбцы со значениями лагов временного ряда
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    # Добавляем столбец со скользящим средним по значению энергопотребления
    # за прошедший rolling_mean_size период времени
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()

# Вызывается функция make_features с параметрами data, max_lag=24, rolling_mean_size=100
make_features(data, 24, 100)

# Создаются обучающая и тестовая выборки с помощью функции train_test_split из библиотеки sklearn. Параметр shuffle=False означает, что данные не перемешиваются, а параметр test_size=0.1 указывает на долю тестовой выборки (10%).
train, test = train_test_split(data, shuffle=False, test_size=0.1)

# Удаляются все строки с пропущенными значениями из обучающей выборки.
train = train.dropna()

# Создается датафрейм X_train, который содержит все столбцы из обучающей выборки train, кроме столбца PJME_MW.
X_train = train.drop('num_orders', axis=1)

# Создается датафрейм y_train, который содержит только столбец PJME_MW из обучающей выборки train.
y_train = train['num_orders']

# Создается датафрейм X_test, который содержит все столбцы из тестовой выборки test, кроме столбца PJME_MW.
X_test = test.drop('num_orders', axis=1)

# Создается датафрейм y_test, который содержит только столбец PJME_MW из тестовой выборки test.
y_test = test['num_orders']
```

```python
# Создаем объект TimeSeriesSplit с заданным количеством фолдов
tscv = TimeSeriesSplit(n_splits=3)
```

```python
# Создаем объект случайного леса
rf = RandomForestRegressor(n_estimators=100, max_depth=10, min_samples_split=2, random_state=42)

# Задаем сетку параметров для поиска лучших гиперпараметров
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [5, 10],
    'min_samples_split': [2, 5],
}

# Используем GridSearchCV для поиска лучших гиперпараметров
grid_rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv=tscv, scoring='neg_mean_squared_error')
grid_rf.fit(X_train, y_train)

# Выводим лучшие гиперпараметры и соответствующий RMSE на кросс-валидации
print("Лучшие гиперпараметры для случайного леса:", grid_rf.best_params_)
print("RMSE на кросс-валидации:", round(np.sqrt(-grid_rf.best_score_),2))
```

```python
# Создаем объект LightGBM регрессора
lgb = LGBMRegressor(n_estimators=100, max_depth=10, learning_rate=0.1, random_state=42)

# Задаем сетку параметров для поиска лучших гиперпараметров
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [5, 10],
    'learning_rate': [0.05, 0.1],
}

# Используем GridSearchCV для поиска лучших гиперпараметров
grid_lgb = GridSearchCV(estimator=lgb, param_grid=param_grid, cv=tscv, scoring='neg_mean_squared_error')
grid_lgb.fit(X_train, y_train)

# Выводим лучшие гиперпараметры и соответствующий RMSE на кросс-валидации
print("Лучшие гиперпараметры для LightGBM:", grid_lgb.best_params_)
print("RMSE на кросс-валидации:", round(np.sqrt(-grid_lgb.best_score_),2))
```

```python
# Создаем объект CatBoostRegressor
catboost = CatBoostRegressor(random_seed=42, verbose=False)

# Задаем сетку параметров для поиска лучших гиперпараметров
param_grid = {
'learning_rate': [0.05, 0.1],
'depth': [5, 10],
'l2_leaf_reg': [1, 3],
}

# Используем GridSearchCV для поиска лучших гиперпараметров
grid_catboost = GridSearchCV(estimator=catboost, param_grid=param_grid, cv=tscv, scoring='neg_mean_squared_error')
grid_catboost.fit(X_train, y_train)

# Выводим лучшие гиперпараметры и соответствующий RMSE на кросс-валидации
print("Лучшие гиперпараметры для CatBoostRegressor:", grid_catboost.best_params_)
print("RMSE на кросс-валидации:", round(np.sqrt(-grid_catboost.best_score_),2))
```

- Я обучил модель случайного леса. Нашёл лучшие гиперпараметры: 'max_depth': 10, 'min_samples_split': 5, 'n_estimators': 100. RMSE на кросс-валидации: 27.85.
- Я обучил модель LightGBM. Нашёл лучшие гиперпараметры: 'learning_rate': 0.1, 'max_depth': 10, 'n_estimators': 50. RMSE на кросс-валидации: 27.51.
- Я обучил модель Catboost. Нашёл лучшие гиперпараметры: 'depth': 5, 'l2_leaf_reg': 1, 'learning_rate': 0.05. RMSE на кросс-валидации: 27.08.

## Тестирование

```python
# Создаем объект модели с гиперпараметрами
model = CatBoostRegressor(depth=5, l2_leaf_reg=1, learning_rate=0.05)

# Обучаем модель на тренировочных данных
model.fit(X_train, y_train)

# Получаем предсказания для тестовых данных
y_pred = model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print('RMSE:', round(rmse,2))
```

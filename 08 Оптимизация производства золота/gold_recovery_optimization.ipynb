### Проверьте, что эффективность обогащения рассчитана правильно. Вычислите её на обучающей выборке для признака rougher.output.recovery. Найдите MAE между вашими расчётами и значением признака. Опишите выводы

```python
# Рассчитываем эффективность обогащения
C = data_train['rougher.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']
rougher_recovery = data_train['rougher.output.recovery']
recovery = (C * (F - T)) / (F * (C - T)) * 100

# Находим MAE
print('MAE =', mean_absolute_error(rougher_recovery, recovery))
```

<div class="alert alert-block alert-info">
<b>КОММЕНТАРИЙ СТУДЕНТА 1:</b><br>
    <ul>
        <li>Результат "9.73512347450521e-15" означает очень маленькое значение, близкое к нулю. Вероятнее всего, это означает, что расчетные значения для эффективности обогащения, рассчитанные по формуле, очень близки к фактическим значениям, которые указаны в столбце 'rougher.output.recovery' в обучающей выборке. Таким образом, можно сделать вывод, что эффективность обогащения, рассчитанная по формуле, была рассчитана правильно.</li>
    </ul> 
</div>

## Модель

### Напишите функцию для вычисления итоговой sMAPE.

```python
def smape(y_true,y_pred):
    # Вычисляем ошибку по формуле SMAPE
    error = np.mean(abs(y_true - y_pred)/((abs(y_true)+abs(y_pred))/2))*100
    # Возвращаем значение ошибки в процентах
    return error

def final_smape(model,rougher_features,rougher_target,final_features,final_target):
    # Вычисляем итоговое значение ошибки на основе SMAPE для признаков rougher.output.recovery и final.output.recovery
    smape_rougher = evaluate_model(model,rougher_features,rougher_target)
    smape_final = evaluate_model(model,final_features,final_target)
    return (0.25*smape_rougher + 0.75*smape_final)
```

<div class="alert alert-block alert-info">
<b>КОММЕНТАРИЙ СТУДЕНТА 1:</b><br>
    <ul>
        <li>Написали функцию smape для измерения среднего абсолютного процентное отклонения sMAPE.</li>
        <li>Написали функцию final_smape для измерения итогового значения ошибки на основе SMAPE для двух признаков: rougher.output.recovery и final.output.recovery.</li>
    </ul> 
</div>

### Обучите разные модели и оцените их качество кросс-валидацией. Выберите лучшую модель и проверьте её на тестовой выборке. Опишите выводы.

```python
# Подготовка данных
common_features = data_train.drop(missing_features, axis=1)

rougher_features = common_features.drop(["date"], axis=1)
rougher_target = data_train["rougher.output.recovery"]

final_features = common_features.drop(["date"], axis=1)
final_target = data_train["final.output.recovery"]

model_dtr = DecisionTreeRegressor(random_state=42)
model_rfr = RandomForestRegressor(n_estimators=10, random_state=42)
model_lr = LinearRegression()

# Создаем функцию-скорер на основе smape
smape_scorer = make_scorer(smape, greater_is_better=False)

def evaluate_model(model, features, target):
    # Оцениваем качество модели кросс-валидацией
    scores = cross_val_score(model, features, target, cv=5, scoring=smape_scorer)
    return -scores.mean()

# Задаем диапазон значений гиперпараметров
max_depth_range = range(1, 11)
min_samples_leaf_range = range(2, 6)

# Итерируемся по всем значениям гиперпараметров и обучаем модель
for max_depth in max_depth_range:
    for min_samples_leaf in min_samples_leaf_range:
        model_dtr = DecisionTreeRegressor(max_depth=max_depth, min_samples_leaf=min_samples_leaf, random_state=42)
        final_smape_dtr = final_smape(model_dtr,rougher_features,rougher_target,final_features,final_target)
        print("Итоговое sMAPE для модели Дерево решений:",round(final_smape_dtr,2),"%", "при максимальной глубине дерева max_depth:",max_depth, "и при минимальном количестве объектов (сэмплов) в листе дерева min_samples_leaf:",min_samples_leaf)

# final_smape_dtr = final_smape(model_dtr,rougher_features,rougher_target,final_features,final_target)
final_smape_rfr = final_smape(model_rfr,rougher_features,rougher_target,final_features,final_target)
final_smape_lr = final_smape(model_lr,rougher_features,rougher_target,final_features,final_target)


print("Итоговое sMAPE для модели Случайный лес:",round(final_smape_rfr,2),"%")
print("Итоговое sMAPE для модели Линейная регрессия:",round(final_smape_lr,2),"%")
```

```python
# создание константной модели, которая предсказывает медиану
dummy = DummyRegressor(strategy='median')
final_smape_dummy = final_smape(dummy,rougher_features,rougher_target,final_features,final_target)
print("Итоговое sMAPE для константной модели Dummy:",round(final_smape_dummy,2),"%")

# сравнение sMAPE для модели Дерева решений и константной модели
model_dtr = DecisionTreeRegressor(max_depth=3, min_samples_leaf=3, random_state=42)
final_smape_dtr = final_smape(model_dtr,rougher_features,rougher_target,final_features,final_target)
if final_smape_dtr < final_smape_dummy:
    print("Модель Дерева решений лучше константной модели")
else:
    print("Константная модель лучше модели Дерева решений")
```
